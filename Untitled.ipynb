```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df_logistic = df[['Survived','Age','SibSp','Parch','Fare']]
```

```python
df_logistic.head()
```

```python
sex = pd.get_dummies(df['Sex'])
df_logistic = pd.concat([df_logistic,sex],axis=1)
df_logistic.columns
```

```python
pclass = pd.get_dummies(df['Pclass'])
df_logistic = pd.concat([df_logistic,pclass],axis=1)
df_logistic.columns
```

```python
embarked = pd.get_dummies(df['Embarked'])
df_logistic = pd.concat([df_logistic,embarked],axis=1)
df_logistic.columns
```

```python
df_logistic.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test = train_test_split(df_logistic.drop('Survived',axis=1),
                                                    df_logistic['Survived'],test_size=0.30,
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test,predictions)
precision = metrics.precision_score(y_test,predictions)
recall = metrics.recall_score(y_test,predictions)
auc = metrics.roc_auc_score(y_test,predictions)
print('Accuracy:',accuracy)
print('Precision:',precision)
print('Recall:',recall)
print('AUC:',auc)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predictions)
```

```python
total = sum(sum(cm))
accuracy = (cm[0,0]+cm[1,1])/total
accuracy
```
